# 1.Slice data with white wine whoes fixed accidity is greater than 5 and less than 9

```python
a1=dataset.loc[(dataset["type"]=="white") & (dataset["fixed acidity"]>5) & (dataset["fixed acidity"]<9),["type","fixed acidity"]]
```

```python
a1
```

# 2. Findout the wine type whoes pH is greater then 3 & less than 3.60 & shows only type & alcohol &pH

```python
a2=dataset.loc[(dataset["pH"]>3) & (dataset["pH"]<3.60),["type","alcohol","pH"]]
```

```python
a2
```

# 3. Findout the wine type whoes critic acid is less than 0.30 and pH is greater than 3.5

```python
a3=dataset.loc[(dataset["citric acid"]<0.30) | (dataset["pH"]>3.5),["type","citric acid","pH"]]
```

```python
a3
```

# 4. Show the quality of wine whoes volatile acidity between is 0.2 to 0.5

```python
a4=dataset.loc[(dataset["volatile acidity"]>0.2) & (dataset["volatile acidity"]<0.5),["type","volatile acidity"]]
```

```python
a4
```

# 5. Shows the data with type fixed acidity, residual sugar, density whoes pH is grater than 3.1 and less than 3.9

```python
a5=dataset.loc[(dataset["pH"]>3.1) & (dataset["pH"]<3.9),["type","fixed acidity","residual sugar","density","pH"]]
```
