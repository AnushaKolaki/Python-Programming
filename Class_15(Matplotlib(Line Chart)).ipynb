# (ms) marker size

```python
a=np.array([1,2,3,4,5])
b=np.array([20,35,22,51,70])
plt.plot(a,b,marker='*',ms=25)
plt.show()
```

# (mec) Marker edge color

```python
a=np.array([1,2,3,4,5])
b=np.array([20,35,22,51,70])
plt.plot(a,b,marker='*',ms=25,mec='y')
plt.show()
```

# (mfc) marker face color

```python
a=np.array([1,2,3,4,5])
b=np.array([20,35,22,51,70])
plt.plot(a,b,marker='*',ms=25,mec='y',mfc='25')
plt.show()
```

# grid

```python
a=np.array([1,2,3,4,5])
b=np.array([20,35,22,51,70])
plt.plot(a,b,marker='*',ms=25,mec='y',mfc='9')
plt.grid()
plt.show()
```

# grid color

```python
a=np.array([1,2,3,4,5])
b=np.array([20,35,22,51,70])
plt.plot(a,b,marker='*',ms=25,mec='y',mfc='9')
plt.grid(color='y')
plt.show()
```

# (ls) line style

```python
a=np.array([1,2,3,4,5])
b=np.array([20,35,22,51,70])
plt.plot(a,b,marker='*',ms=25,mec='y',mfc='9',ls='--')
plt.grid(color='y')
plt.show()
```

# grid line style(ls)
