```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
a=sns.load_dataset("tips")
a
```

```python
sns.set(style="whitegrid")
```

```python
sns.relplot(data=a,x="total_bill",y="sex")
```

```python
sns.relplot(data=a,x="total_bill",y="tip")
```

```python
sns.relplot(data=a,x="total_bill",y="tip",color="r",style="sex")
```

```python
sns.relplot(data=a,x="total_bill",y="tip",style="day")
```

```python
sns.set(style="darkgrid")
```

```python
sns.relplot(data=a,x="total_bill",y="tip",size="day")
```
